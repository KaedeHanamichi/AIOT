```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display  # 用於在 Colab 中顯示圖片

# 載入模型
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # 使用 GPU 加速

# 自定義函數來生成圖片
def generate_image(prompt):
    """
    根據輸入的文字描述生成影像，並在 Colab 中顯示。
    :param prompt: 用於描述影像的文字。
    """
    print(f"Generating image for: {prompt}")
    image = pipe(prompt).images[0]  # 生成影像
    display(image)  # 在 Colab 中顯示圖片
    return image

# 輸入文字並生成影像
text_prompt = input("請輸入你的文字描述: ")
image = generate_image(text_prompt)
```
